Question 3: Write a python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point.

```python
import numpy as np;
import matplotlib.pyplot as plt;
# Creating a series of data of in range 1-50.
x = np.linspace(1,50,200);
mean= np.mean(x);
sd = np.std(x);
factor = (sd*((2*np.pi)**(1/2)));
prob_density = ((1/factor)*(np.exp(-0.5*((x-mean)/sd)**2)));

plt.plot(x, prob_density, color = 'blue');
plt.xlabel('Data Points');
plt.ylabel('Probability Density');
```

Question 7: How Binomial Distribution different from poissson distribution?

```python
Answer_7 = """ 
The key differences between the two distributions are:

Nature of the events: In the Binomial distribution, the events are discrete and have a binary outcome (success or failure),
whereas in the Poisson distribution, the events are rare occurrences that can be considered continuous.

Number of trials/samples: The Binomial distribution models the number of successes in a fixed number of trials or samples,
whereas the Poisson distribution models the number of rare events that occur in a fixed interval or sample size.

Probability of success/rate of occurrence: In the Binomial distribution, the probability of success is constant across trials or samples,
whereas in the Poisson distribution, the rate of occurrence is constant over time or space.

Assumptions: The Binomial distribution assumes that the trials are independent and identically distributed,
whereas the Poisson distribution assumes that the rare events occur independently of each other and that the rate of occurrence is constant over time or space.
"""
```

Question 8: Generate a random sample of size 1000 from a poisson distribution and poisson distribution?
