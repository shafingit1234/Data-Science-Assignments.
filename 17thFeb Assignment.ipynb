```python
# What is MongoDB?Explain non-relational databases in short. In which scenarios it is preferred to use Mongodb over sql databases.
ans = """ 
    MongoDB is a document-oriented NO-SQL database used of high-volume data storage.
    It is different from relational databases since we don't store data in the form of rows or tables.
    Here we make use of collections and documents, where documents stores data in json like format or in key-value pairs.
    Collections on the other hand contains the set of documents.
    
    Non_relational Databases:
    Non-relational databases are different from traditional relational databases, since here idea of storing data is simplified.
    Rather than following strict convention of first forming schema then storing data based on that schema,
     non-relational databases allows us to store data in key-value form or json-like form.
    Instead of forming tables, we here form collections and documents.
    The ability to digest and organize various types of information side by side makes non-relational databases much more flexible than relational databases.
    
    Mongodb is designed for horizontal scalability, making it a good choice for applications with rapidly growing amounts of data,
    in such a case where it is certain that a day will come when company's product will cater to a million data, it is better to use mongoDB.
    
"""
```

```python
# State and explain the features of MongoDB.
ans  = """ 
1. Schema-less database:- We don't need to worry about schema formation since we can store data in the from key-value pair.
2. Document-oriented:- All the data are stored in documents instead of tables like RDBMS.
2. Indexing:- Every field here is indexed with both primary and secondary indices which makes easier and efficient searches.
3. Scalability:- It provides horizontal scalability with the help of sharding. Sharding means to distribute data on multiple servers.
4. Replication:- With the help of replication it creates, multiple copies of the data and sends these copies to a different server,
                 so that if one server fails, then the data is retrieved from another server.
5. Aggregation:- It allows to perform operations on the grouped data and get a single result or computed result.
6. High Performance:- The performance of MongoDB is very high and data persistence as compared to another database,
                      due to its features like scalability, indexing, replication etc.
"""
```
