```python
# Question3: What is app routing in flask, why do we use app routes?
ans=""" 
    App routing means mapping the urls to a specific functin that will handle the logic for that url.
    Modern web frameworks use more meaningful URLs to help users remember the URLs and make navigation simpler.
    
    Routes in flask are mapped to python functions. The route() decorator binds a url to a function.
    Basically routes are used for url to function binding.
    Example, suppose I want to bind url /hello_world with a function that displays hello world on screen.
    For that I can use below written code to acheive the desired functionality.
    @app.route("/hello_world")
    def display():
        return "hello world";
    
    When you will hit the url '/hello_world' then route will help in executing the required function.
"""
```

```python
# What function is used in flask for url building? Write a python code to demonstrate the working of the url_for() function.
ans = """ 
    url_for is used in flask for url building.
    url_for() function is used to build a URL to the specific function dynamically.
    The first argument is the name of the specified function,
    and then we can pass any number of keyword argument corresponding ot the variable part of the url.
    In below example, I have demonstrate how url_for comes handy.
    I have defined api's like /student that will display student name,
    /course:- display course_name, /mentor:- display mentor name.
    I have also defined other api that takes url in the form of /user/<name>,
    where name can be anything, and this api is binded to a user function,
    which checks if name = 'mentor' redirect to name_of_mentor() function using url_for() method,
    Similiarly if name = 'student' redirect to name_of_student() function using url_for() method,
    Similiarly if name = 'course'  redirect to name_of_course()  function using url_for() method,
    if name = 'anything' then it will be redirected to a default function that will print could not find the page.
"""
code = """ 
from flask import Flask;
from flask import request as req;
from flask import redirect;
from flask import url_for;
app = Flask(__name__);

@app.route('/user/<name>')  
def user(name):  
    if name == 'mentor':  
        return redirect(url_for('name_of_mentor'))  
    if name == 'student':  
        return redirect(url_for('name_of_student'))  
    if name == 'course':  
        return redirect(url_for('name_of_course')) 
    if name == 'feedback':
        return redirect(url_for('feedback'));
    return redirect(url_for('default'));


@app.route("/mentor")
def name_of_mentor():
    return "<h1>Mentor Name: Sudhanshu Kumar</h1>";

@app.route("/student")
def name_of_student():
    return "<h1>Student Name: Shafin</h1>";
@app.route("/course")
def name_of_course():
    return "<h1>Course Name: Data Science</h1>";

@app.route("/feedback")
def feedback():
    return "<h1>Enjoying learning rest apis</h1>";
@app.route("/default")
def default():
    return "Couldn't find the page you want to display";

if(__name__ == "__main__"):
    app.run(host="0.0.0.0");


here if I will call api suppose /user/course then after hiting enter button, url will be redirected to /course api automatically,
however I have redirected to /course api functions using url_for() method.
"""
print("screenshot for API called:- /user/course");
print("*************************************");
display.Image('https://raw.githubusercontent.com/shafingit1234/Data-Science-Assignments./main/course_name.png')
```
