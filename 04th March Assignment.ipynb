```python
# Load the titanic dataset using the load_dataset function of seaborn, use plotly express to plot a scatter plot for age and fare columns in the titanic dataset.
import seaborn as sns
import plotly.express as px

# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")
# td = px.data.titanic();
# print(titanic_data);
# Create a scatter plot for age and fare columns
scatter_plot = px.scatter(titanic_data, x="age", y="fare", color = 'survived' ,size = 'fare')

# Show the scatter plot
scatter_plot.show()
```

```python
# Using the tips dataset in the plotly library, plot a boxplot using plotly express?
tips_data = px.data.tips();
# print(tips_data);
box_plot= px.box(tips_data,y="total_bill");
box_plot.show();
```

```python
# Using the tips dataset in the plotly library, plot a histogram for x = "sex" and y = "total_bill" col in the tips dataset. Also use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.
tips_data = px.data.tips()
histogram = px.histogram(tips_data, x="sex", y="total_bill", color="day", pattern_shape="smoker")
histogram.show()
```

```python
# Using the iris dataset in the plotly library. Plot a scatter matrix plot, using the "species" column for the color parameter?
iris_data = px.data.iris()
scatter_matrix = px.scatter_matrix(
    iris_data,
    dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
    color="species"
)
scatter_matrix.show()
```
