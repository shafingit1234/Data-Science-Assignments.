```python
# Why is redirect() used in flask?
ans = """ 
    Flask redirect is defined as a function in flask which allows developer to redirect users to a specified URL.
    When this function is called, a response object is returned, and the redirection happens to the target location.
    Example:
    from flast import redirect;
    app = Flask(__name__);
    @app.route("/")
    def home():
        return "Welcome to my home";
    @app.route("/redirectOne")
    def redirectOne():
        return redirect("/");
    @app.route("/redirectTwo")
    def redirectTwo():
        return redirect("/landing");
    @app.route("/landing")
    def landing():
        return "Landing Hello"
    
    if(__name__ == '__main__'):
        app.run(host="0.0.0.0" , port = 8080);
    
    In above example, when api /redirectOne is entered, we will be redirected to "/" api or home() function.
    When api /redirectTwo is called, then we will redirected to "/landing" api or landing() function.
    
"""
```
