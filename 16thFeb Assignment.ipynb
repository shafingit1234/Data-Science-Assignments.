```python
# What is DDL? Explain why create, drop, alter and truncate are used with an example?
ans = """
DDL(Data Definition Language):- It is a language used to create and modify the structure of database objects in a database,
using predefined commands and a specific syntax.
It is a set of SQL commands responsible in creating description or definition of database.
List of DDL commands:
1. Create.: This command is used to create the database or its objects (like table, index, function, views, store, procedure and triggers)
2. Drop.: This command is used to delete objects from the database.
3. Alter.: This is used to alter the structure of the database.
4. Truncate.: This is used to remove all records from a table, including all spaces allocated fro the records are removed.

To create table schema or table we use CREATE command
example:
CREATE TABLE pwskills(
    courses varchar(30),
    price number,
    course_id number PRIMARY KEY,
)
To Drop objects we make use of drop command.
Example:
drop table pwskills;

Alter: example
Alter table pwskills add course_head_mentor varchar(40);
//It will add another coloumn named course_head_mentor in pwskills table.

Truncate example:
TRUNCATE TABLE pwskills;
"""
```

```python
# What is DML? Explain insert, update and delete with an example?
ans = """ 
DML(Data Manipulation Language):- 
Data manipulation language is a set of sql commands that helps in manipulating data stored in our database table.
DML provides us a way to add, update, fetch or delete data from the database.
Total dml commands present in sql are:=

1. INSERT: This command allows users to insert data in database tables.
2. UPDATE: This command allows users to update or modify the existing data in database tables.
3. DELETE: Allows users to remove single or multiple existing records from the database tables.

Insert example:
insert into pwskills values('Data Science' , 3500, 123400 , "Sudhanshu Kumar");

Update example:
update pwskills set price = 3200 where courses = 'Data Science';

Delete Example:
Delete from pwskills where course_id = 123400;

"""
```
