```python
# Question: Write a python program for encrypting a message sent to you by your friend.
def convert(item):
    punctuation = ['.', '!' , ','];
    new_word = "";
    for i in item:
        if i in punctuation:
            new_word = new_word + i;
            continue;
        j = ord(i) - 97;
        new_word = new_word + chr(25 - j + 97);
    return new_word;

def main():
    orig = "I want to become a Data Scientist.";
    sample_sentence = ("I want to become a Data Scientist.").casefold();
    # white space to be replaced by $,
    # a to be replaced by $;
    # b to replaced by y, c replaced by x so on.
    # for i in sample_sentence:
    ans = "";
    end = len(sample_sentence);
    i=0;
    while i <end:
        idx = sample_sentence.find(" ", i, end);
        if idx == -1:
            idx = end;
        word_to_replace = sample_sentence[i:idx];
        i = idx+1;
        new_word = convert(word_to_replace);
        # print(new_word);
        sample_sentence = sample_sentence.replace(word_to_replace, new_word);
    # print(sample_sentence);
    print("Original Sentence goes here:-");
    print("----------------------------------------------");
    print(orig);
    print("----------------------------------------------");
    print("Encrypted sentence goes here:-");
    print("----------------------------------------------");
    print(sample_sentence);

main();
        
    
    
```
