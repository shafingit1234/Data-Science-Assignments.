```python
# You have a Pandas DataFrame df that contains a column named 'Email' that contains Email addresses in the format 'username@domain.com'. Write a python function that creates a new column 'Username' in df that contains only the username part of each email address.
import random;
import string;
email_ids = list();
for i in range(0,12):
    kprime = random.randint(3,9);
    email_id = ''.join(random.choices(string.ascii_lowercase, k=kprime));
    email_id = email_id + "@domain.com";
    email_ids.append(email_id);
new_data = {'Email-Id' : email_ids};
df = pd.DataFrame(new_data);
print("*************** Original Data Goes below *******");
print(df);
df['username']  = "";
for i, j in df.iterrows():
    idx = j['Email-Id'].find('@');
    sub = j['Email-Id'][0:idx];
    j['username'] = sub;
print("*************** Updated Data Goes below ********");
print(df);
```

```python
# You have a pandas DataFrame df with a column 'Date'. Write a pythoh function that creates a new column 'Weekday' in the Dataframe. The 'Weekday' column should contain the weekday name corresponding to each date in the 'Date' column.
import datetime;
def findDate(year, month, date):
    idx =int(month);
    months = ['January' , 'Feburary' , 'March' , 'APril' , 'May' , 'June' , 'July' , 'August' , 'September', 'October' , 'November' , 'December'];
    dt =int(date);
    res = months[idx-1] + " " + str(dt) + ", " + year;
    return res;

df2 = pd.DataFrame();
df2['Date'] = df['date'];
# print(df2);
print("***************** Original Data ****************");
print(df2);
df2['Day'] = '';
for i, j in df2.iterrows():
    res = j['Date'].split('-');
    year = res[0];
    month= res[1];
    date = res[2];
    dt = findDate(year, month, date);
    j['Day'] = datetime.datetime.strptime(dt , '%B %d, %Y').strftime('%A');
print("**************** Updated Data *****************");
print(df2);
```
